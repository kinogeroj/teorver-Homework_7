```python
import numpy as np
import scipy.stats as stats


# 1. Даны две  независимые выборки.
# Не соблюдается условие нормальности.
# x1: 380, 420, 290
# y1: 140, 360, 200, 900
# Сделайте вывод по результатам, полученным с помощью функции.

x1_data = np.array([380,420,290])
y1_data = np.array([140,360,200,900])

alpha = 0.05

print(f'Результат оценки {stats.mannwhitneyu(x1_data, y1_data)}, pvalue > {alpha}! Статистически значимых различий нет. Следовательно, принимаем нулевую гипотезу.')
```

```python
# 2. Исследовалось влияние препарата на уровень давления пациентов.
# Сначала измерялось давление до приема препарата,
# потом через 10 минут и через 30 минут. 
# Есть ли статистически значимые различия?
# 1е измерение до приема препарата: 150, 160, 165, 145, 155.
# 2е измерение через 10 минут: 140, 155, 150,  130, 135.
# 3е измерение через 30 минут: 130, 130, 120, 130, 125.

x1_data = np.array([150, 160, 165, 145, 155])
x2_data = np.array([140, 155, 150, 130, 135])
x3_data = np.array([130, 130, 120, 130, 125])

alpha = 0.05

print(f'Результат оценки {stats.friedmanchisquare(x1_data, x2_data, x3_data)}, pvalue < {alpha}! Статистически значимые различия есть. Следовательно, принимаем альтернативную гипотезу.')
```

```python
# 3. Сравните 1 и 2-е измерения, предполагая, 
# что 3-го измерения через 30 минут не было.

x1_data = np.array([150, 160, 165, 145, 155])
x2_data = np.array([140, 155, 150, 130, 135])

alpha = 0.05

print(f'Результат оценки {stats.wilcoxon(x1_data, x2_data)} pvalue > {alpha}! Статистически значимых различий нет. Следовательно, принимаем нулевую гипотезу.')
```

```python
# 4. Даны 3 группы  учеников плавания.
# В 1 группе время на дистанцию 50 м 
# составляют: 56, 60, 62, 55, 71, 67, 59, 58, 64, 67.
# Вторая группа: 57, 58, 69, 48, 72, 70, 68, 71, 50, 53.
# Третья группа: 57, 67, 49, 48, 47, 55, 66, 51, 54.

x1_data = np.array([56, 60, 62, 55, 71, 67, 59, 58, 64, 67])
x2_data = np.array([57, 58, 69, 48, 72, 70, 68, 71, 50, 53])
x3_data = np.array([57, 67, 49, 48, 47, 55, 66, 51, 54])

alpha = 0.05

print(f'Результат оценки {stats.kruskal(x1_data, x2_data, x3_data)} pvalue > {alpha}! Статистически значимых различий нет. Следовательно, принимаем нулевую гипотезу.')
```

```python
# 5. Заявляется, что партия изготавливается со средним
# арифметическим 2,5 см. 
# Проверить данную гипотезу, если известно, 
# что размеры изделий подчинены нормальному закону распределения.
# Объем выборки 10, уровень статистической значимости 5%.
# 2.51, 2.35, 2.74, 2.56, 2.40, 2.36, 2.65, 2.7, 2.67, 2.34.

data = np.array([2.51, 2.35, 2.74, 2.56, 2.40, 2.36, 2.65, 2.7, 2.67, 2.34])
n = 10

M = data.mean() # Среднее значение по измерениям

std = data.std(ddof = 1)

# Используем формулу для критерия Стьюдента. Нормальное распределение, дисперсия неизвестна.

t = (M - 2.5) / (std / np.sqrt(n))

stats.t.ppf(0.975, 9) # 9 степеней свободы (n - 1)

print(f'Наше расчетное значение {stats.t.ppf(0.975, 9)} > наблюдаемого {t}! Следовательно, при заданном уровне статистической значимости, статистически значимых различий нет. Принимаем нулевую гипотезу.')
```
